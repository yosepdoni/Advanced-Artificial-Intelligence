```
import pytholog as pl

siswa_kb = pl.KnowledgeBase("siswa")

siswa_kb([
    "tugas(andi, 80)",
    "tugas(beni, 50)",
    "tugas(citra, 30)",
    "jumlah_ekskul(andi, 2)",
    "jumlah_ekskul(beni, 1)",
    "jumlah_ekskul(citra, 3)",
    "stres(X, P) :- tugas(X, T), P is T / 100",
    "disiplin(X, 0.8) :- jumlah_ekskul(X, J), J >= 1",
    "motivasi(X, 0.7) :- jumlah_ekskul(X, J), J >= 1",

    # aturan kelelahan
    "lelah(X, 0.3) :- jumlah_ekskul(X, J), J >= 2",

    # jika hanya 1 ekskul, lelah rendah
    "lelah(X, 0.1) :- jumlah_ekskul(X, 1)",

    # prestasi akhir
    "prestasi(X, P) :- disiplin(X, D), motivasi(X, M), lelah(X, L), stres(X, S), P is D + M - L - (S * 0.5)"
])

print(siswa_kb.query(pl.Expr("stres(Nama, Nilai)")))
print(siswa_kb.query(pl.Expr("prestasi(Nama, Nilai)")))
```
